## Filtering data

```python
df
```

```python
df.loc[(df['work_year'] == 2023) & (df['salary'] == 80000)] #using multiple filters
```

```python
new_df = df.loc[(df['work_year'] <= 2023) & (df['salary'] == 80000)] 
new_df
```

```python
new_df.reset_index(drop=True, inplace=True) #reset index and remove old index
new_df
```

## Regex Filtering

```python
analyst = df.loc[df['job_title'].str.contains('Analyst')] 
analyst.reset_index(drop=True, inplace=True)
analyst.to_csv('Analysts_data.csv')
```

```python
import re #importing regular expression
engineer = df.loc[df['job_title'].str.contains('ai|engineer', flags=re.I, regex=True)] #using regex #flags to omit capitalisation
engineer
```

## Conditional changes

```python
df.loc[df['experience_level'] == 'MI', 'experience_level'] = 'Mid/Intermediate level' #changing values inside dataset
df
```

```python
df.loc[df['experience_level'] == 'SE', 'experience_level'] = 'Senior'
df.loc[df['experience_level'] == 'EN', 'experience_level'] = 'Entry level'
df.loc[df['experience_level'] == 'EX', 'experience_level'] = 'Executive level'
df.loc[df['experience_level'] == 'FL', 'experience_level'] = 'Freelancing'
df.loc[df['experience_level'] == 'CT', 'experience_level'] = 'Contractor'
df
```

```python
df.head()
```

```python
df.loc[df['salary_in_usd'] >= 80000, ['salary']] = 'Good Pay' 
df
```
