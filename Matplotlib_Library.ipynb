```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

## Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#plt.plot(x,y, label='2x', color='red', linewidth = 2, linestyle='--', marker='.', markersize=10, markeredgecolor='blue')

#Resize the graph
plt.figure(figsize=(5,4), dpi=300)

#Shorthand notation {color}{marker}{line}
plt.plot(x,y,'g.--', label='2x', markeredgecolor='blue')


#Adding font type and font size
plt.title('Basic Graph', fontdict={'fontname':'Times New Roman', 'fontsize':15}) 
plt.xlabel('X-Axis', fontdict={'fontname':'Times New Roman', 'fontsize':10})
plt.ylabel('Y-Axis', fontdict={'fontname':'Times New Roman', 'fontsize':10})

#Line Number two
x2 = np.arange(0,4,0.5)
plt.plot(x2[:5], x2[:5]**2, 'r', label ='Squared x')
plt.plot(x2[4:], x2[4:]**2, 'r--')


#To add value to axises
plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,1,2,3,4,5,6,7,8])


#Adding legend
plt.legend()

#Saving graph
#plt.savefig('basicgraph.png', dpi=300)

plt.show()
```

## Bar Chart

```python
labels = ['A','B','C']
values=[1,3,4]

plt.figure(figsize=(5,3), dpi=150)

plt.title('Bar Chart')

bars = plt.bar(labels,values)

patterns=['/','.','-']
for i in bars:
     i.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('.')
#bars[2].set_hatch('-')

plt.show()
```

## 1. Gas Price Dataset

```python
#Importing libraries
import numpy as np
import pandas as pd

#Working with gas prices dataset
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

```python
gas.info()
```

```python
gas.set_index('Year')
```

```python
#Resizing the chart
plt.figure(figsize=(8,5), dpi=300)

#Plotting based on year
plt.plot(gas.Year, gas.USA, 'r.-', label='USA')
plt.plot(gas.Year, gas.Canada,'b.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], label='South Korea', marker='.', markeredgecolor='yellow' )

#Adding additional info
plt.title('Gas Price Chart', fontdict={'fontname':'Times New Roman', 'fontsize' : 20})
plt.xlabel('Year',fontdict={'fontname':'Times New Roman', 'fontsize' : 15})
plt.ylabel('US Dollars', fontdict={'fontname':'Times New Roman', 'fontsize' : 15})

#Adding values to ticks
plt.xticks(gas.Year[::3])

#Adding legend
plt.legend()

plt.show()
```

## Doing the exact same for all countries

```python
#Resizing the chart
plt.figure(figsize=(10,7), dpi=500)

#Plotting based on year
#plt.plot(gas.Year, gas.USA, 'r.-', label='USA')
#plt.plot(gas.Year, gas.Canada,'b.-', label='Canada')
#plt.plot(gas.Year, gas['South Korea'], label='South Korea', marker='.', markeredgecolor='yellow' )

#Plotting for all countries in list
#countries_looking = ['South Korea', 'UK', 'Japan','France','Germany'] #To look for specific countries
for country in gas:
    if country != 'Year':
   #if country in countries_looking: 
        plt.plot(gas.Year, gas[country], marker ='.', label = country)

#Adding additional info
plt.title('Gas Price Chart', fontdict={'fontname':'Times New Roman', 'fontsize' : 20})
plt.xlabel('Year',fontdict={'fontname':'Times New Roman', 'fontsize' : 15})
plt.ylabel('US Dollars', fontdict={'fontname':'Times New Roman', 'fontsize' : 15})

#Adding values to ticks
plt.xticks(gas.Year[::3])

#Saving the chart
plt.savefig('Gas_Price_Chart.png', dpi=500)

#Adding legend
plt.legend()

plt.show()
```

## Doing the exact same for specific countries in list

```python
#Resizing the chart
plt.figure(figsize=(10,7), dpi=500)

#Plotting based on year
#plt.plot(gas.Year, gas.USA, 'r.-', label='USA')
#plt.plot(gas.Year, gas.Canada,'b.-', label='Canada')
#plt.plot(gas.Year, gas['South Korea'], label='South Korea', marker='.', markeredgecolor='yellow' )

#Plotting for all countries in list
countries_looking = ['South Korea', 'UK', 'Japan','France','Germany'] #To look for specific countries
for country in gas:
    if country in countries_looking:  
        plt.plot(gas.Year, gas[country], marker ='.', label = country)

#Adding additional info
plt.title('Gas Price Chart', fontdict={'fontname':'Times New Roman', 'fontsize' : 20})
plt.xlabel('Year',fontdict={'fontname':'Times New Roman', 'fontsize' : 15})
plt.ylabel('US Dollars', fontdict={'fontname':'Times New Roman', 'fontsize' : 15})

#Adding values to ticks
plt.xticks(gas.Year[::3].tolist()+[2011])

#Adding legend
plt.legend()

plt.show()
```

## 2. Fifa Dataset

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
fifa.info()
```

```python
fifa.set_index('ID')
```

## (i) Histogram

```python
#Creating bins
bins = [40,50,60,70,80,90,100]

#Adding title and customising
plt.title('Distribution of player skills in FIFA 2018')

plt.ylabel('Number of players')
plt.xlabel('Skill level')

plt.xticks(bins)

#Histogram
plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.show()
```

## (ii) Pie Charts (Preferred Foot)

```python
fifa.columns
```

```python
fifa['Preferred Foot'].count()
```

```python
#Counting the right and left foot
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]

#Labelling and customising
labels = ['Right foot','Left Foot']
colors = ['#7c89fc', '#e57cfc']

#Pie Chart
plt.pie([right,left], labels = labels, colors = colors, autopct = '%.2f %%')

#Chart name
plt.title('PREFERRED FOOT', fontdict={'fontname' : 'Times New Roman', 'fontsize' : 15})

#Adding the legends
plt.legend()

plt.show()
```

## (iii) Pie Charts (Player weights)

```python
fifa.Weight.count()
```

```python
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]

fifa.Weight[0]
```

```python
#Type of players (Classification)
light_players = fifa.loc[fifa.Weight < 125].count()[0]
light_medium_players = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]

medium_players = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy_players = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]

heavy_players = fifa.loc[fifa.Weight >= 200].count()[0]

#Weight list of all players
weights = [light_players, light_medium_players, medium_players, medium_heavy_players, heavy_players]

#Adding title and customising
plt.title('Weight Distribution (in lbs)', fontdict={'fontname': 'Times New Roman', 'fontsize': 20 })

labels = ('under 125', '125-150','150-175', '175-200', 'Over 200')
explode = (.1, .1, .1, .1, .1)

#pie Chart
plt.pie(weights, labels=labels, autopct = '%.2f %%', pctdistance = 0.8, explode =explode)
#https://stackoverflow.com/questions/50602158/matplotlib-pie-chart-variable-pctdistance


plt.show()
```
